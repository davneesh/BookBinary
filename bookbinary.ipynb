# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name.info()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean('Book-Rating')['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author_x','Image-URL-M_x','num_ratings','avg_rating']]
```

```python
popular_df
```

# Colloborative Filtering

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pvt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pvt.shape
```

```python
pvt.fillna(0,inplace=True)
```

```python
pvt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pvt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  index = np.where(pvt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pvt.index[i[0]])
```

```python
recommend('1984')
```
